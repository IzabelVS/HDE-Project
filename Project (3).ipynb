**Task 1.2**

We derive $u, u_{tt},$ & $u_{xx}$ for:
$$
u_{tt} + \nu u_{t} = c^2 u_{xx}
$$

We find $u_{xx}$, for which we can derive in the sum as $n$ is just a constant:
$$
u(x, t)_{x} = \frac{\partial\sum_{n=1}^{\infty} f_n(t) \sin(\lambda_n x)}{\partial x}=\sum_{n=1}^{\infty} f_n(t)\frac{d \sin(\lambda_n x)}{dx}  =\sum_{n=1}^{\infty} f_n (t) \lambda_n\cos(\lambda_n x) 
$$
$$
u(x, t)_{xx} = \frac{\partial f_n (t) \lambda_n\cos(\lambda_n x)}{\partial x}=\sum_{n=1}^{\infty} f_n(t)\lambda_n\frac{d \cos(\lambda_n x)}{dx}  =\sum_{n=1}^{\infty} f_n (t) \lambda_n-\lambda_n\sin(\lambda_n x)=-\lambda_n^2\sum_{n=1}^{\infty} f_n (t) \sin(\lambda_n x) 
$$

We can derive for $u_t$ and $u_{tt}$ similarly:
$$
u(x, t)_{t} = \frac{\partial\sum_{n=1}^{\infty} f_n(t) \sin(\lambda_n x)}{\partial t}=\sum_{n=1}^{\infty} \frac{d f_n(t)}{d t} \sin(\lambda_n x) =\sum_{n=1}^{\infty} f_n '(t) \sin(\lambda_n x) 
$$
By this same logic:
$$
u(x, t)_{tt} = \frac{\partial\sum_{n=1}^{\infty} f_n '(t) \sin(\lambda_n x) }{\partial t}=\sum_{n=1}^{\infty} f_n ''(t)\sin(\lambda_n x) 
$$


Substituting into the PDE:
$$
\sum_{n=1}^{\infty} f_n ''(t)\sin(\lambda_n x) + \nu \sum_{n=1}^{\infty} f_n '(t) \sin(\lambda_n x)  = -c^2 \lambda_n^2\sum_{n=1}^{\infty} f_n (t) \sin(\lambda_n x) 
$$

Re-arranging and putting these sums together:
$$
\sum_{n=1}^{\infty} ( f_n ''(t)\sin(\lambda_n x) + \nu f_n '(t) \sin(\lambda_n x)  + (c\lambda_n)^2 f_n (t) \sin(\lambda_n x) ) = 0
$$

$$
\sum_{n=1}^{\infty} \sin(\lambda_n x)( f_n ''(t) + \nu f_n '(t)  + (c\lambda_n)^2 f_n (t)) = 0
$$

As for some n $\sin(\lambda_n x)\neq0$, so we disregard it, which gets us the following ODE for $n = 1, 2, \dots, N $:
$$
 f_n ''(t) + \nu f_n '(t)  + (c\lambda_n)^2 f_n (t) = 0
$$

The Fourier series of f(x) is given by:

f(x) = \frac{a_0}{2} + \sum_{n=1}^\infty ( a_n cos(\frac{n\pi x}{L}) + b_n cos(\frac{n\pi x}{L}))

$f_n(0)$ is the $b_n$ coefficient of the Fourier expansion. Therefore, by the Euler-Fourier formula for $b_n$ the intial condition of $f_n$ is:

$$
f_n(0) =  \frac{2}{L} \int_{0}^{L} f(x) \sin(\frac{n\pi x}{L}) \,dx = \frac{2}{L} \int_{0}^{L} (\exp\left(\sin^2\left(\frac{2\pi x}{L}\right)\right)-1) \sin(\frac{n\pi x}{L}) \,dx
$$

Similarly using $u_t\left(x, t=0\right) = 0$:
$$
0 = u_t\left(x, t=0\right)  = \sum_{n=1}^{\infty} f_n' (0) \sin(\lambda_n x) 
$$
The Fourier series of $u_t(x)$ is given by:
$$
u_t(x) = \frac{a_0}{2} + \sum_{n=1}^\infty ( a_n cos(\frac{n\pi x}{L}) + b_n cos(\frac{n\pi x}{L}))
$$

$f_n'(0)$ is the $b_n$ coefficient of this Fourier expansion. Therefore, by the Euler-Fourier formula for $b_n$ the intial condition of $f_n'$ is:

$$
f_n'(0) =  \frac{2}{L} \int_{0}^{L} u_t(x) \sin(\frac{n\pi x}{L}) \,dx = \frac{2}{L} \int_{0}^{L} 0 \sin(\frac{n\pi x}{L}) \,dx = 0

```python
#Import modules
import sympy as sym
import numpy as np

def calc_fn_0(k):
    '''
    Calculates ùëì‚Çñ(0)  using sympy for a function with initial condition:
        ùë¢(ùë•,ùë°=0)=exp(sin2(2ùúãùë•ùêø))‚àí1. 
    The integrand has its period lengthened by a factor of L. This increases its 
    integral over its period by a factor of L. However this integral is divided 
    by L, therefore L has no effect on the result and we can arbitrarily  it to 2.
    -------------------------------------------------------
    Input: 
        ‚Ä¢ k - int; the index of the terms in the series solution
    Output:
        ‚Ä¢ ùëì‚Çñ(0) - float
    '''
    #Set up x symbol
    x = sym.symbols('x')
    
    #Define integrand
    igrd = (sym.exp(sym.sin(sym.pi*x)**2) - 1) * sym.sin(k*sym.pi*x/2)
    
    #Evaluate the integrand
    return sym.integrate(igrd, (x, 0, 2)).evalf()

#Finds f‚Çô(0)
print(calc_fn_0(5))
```

Define new variables:

$$Fn = \begin{bmatrix}F_{n,1}\\ F_{n,2}\end{bmatrix}; F_{n,1} = f_n(t); F_{n,2} = f_n'(t)$$

Based on $f_n ''(t) + \nu f_n '(t)  + (c\lambda_n)^2 f_n (t) = 0$:
$$F_{n,1}'= F_{n,2}; F_{n,2}'= f_n''(t) = -\nu f' - (c\lambda_n)^2 f = -\nu  F_{n,2} - (c\lambda_n)^2 F_{n,1}$$

$$Fn' = 
\begin{bmatrix}
F_{n,1}'\\ F_{n,2}'
\end{bmatrix}
=\begin{bmatrix}F_{n,2}\\ -\nu  F_{n,2} - (c\lambda_n)^2 F_{n,1}\end{bmatrix}$$

```python
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
    
def dF_dt(F, t, k):
    '''
    This function defines the RHS of the system of ODEs for f‚Çô's
    ----------------------------------------------------
    Inputs: 
        ‚Ä¢ F - array; contains [f‚Çô, f‚Çô']
        ‚Ä¢ t - float; time 
        ‚Ä¢ k - int; the index of the terms in the series solution
    Output
        ‚Ä¢ array; RHS of the system of ODEs
    '''
    fn, fn_prime = F[0], F[1]
    
    #Finds each dF/dt
    return [fn_prime, -nu * fn_prime - (c * (k * np.pi / 10) ** 2) * fn]

def fn_solve(k, t_vec):
    '''
    This function derives the solution of f_k(t) for all the times in 't_vec'
    'k' is given as an input and marks the index of the term in series solution 
    --------------------------------------------------------------------------
    Inputs: 
        ‚Ä¢ k - int; the index of the terms in the series solution
        ‚Ä¢ t_vec - array; vector of times tested
    Output:
        ‚Ä¢ fk_t - array; solution at each point in t_vec
    '''    
    #Initial conditions based on previous derivations
    intl_cond = [calc_fn_0(k), 0]
    
    #Solves ODE
    sol = odeint(dF_dt, intl_cond, t_vec, args=(k,))

    return sol[:, 0]


#Define constants
nu = .1
c = 1
L = 10

#Define t tested
t_vec = np.linspace(0, 100, 1001)

#Define n values
n_val = [1, 2, 3]

#Loops for each n
for n in n_val:
    #Solves for f‚Çô and plots
    fnsol = fn_solve(n, t_vec)
    plt.plot(t_vec, fnsol, label='f'+str(n))

# PLOT AESTHETICS
#Grid
plt.grid(True, linestyle='-')    

plt.legend()

#Labels axis and title
plt.xlabel('t')
plt.title('Solutions for f‚Çô(t)')

plt.show()
```

Using $u(x, t) \approx \sum_{n=1}^{N} f_n(t) \sin(\lambda_n x)$, we construct this matrix of solutions at times 1 to N$_t$ and points 1 to N$_x$:
$$
\begin{bmatrix}
\sum_{n=1}^{N} f_n(t_1) \sin(\lambda_n x_1) & \sum_{n=1}^{N} f_n(t_2) \sin(\lambda_n x_1) & \dots & \sum_{n=1}^{N} f_n(t_{N_t}) \sin(\lambda_n x_1) \\
\sum_{n=1}^{N} f_n(t_1) \sin(\lambda_n x_2) & \sum_{n=1}^{N} f_n(t_2) \sin(\lambda_n x_2) & \dots & \sum_{n=1}^{N} f_n(t_{N_t}) \sin(\lambda_n x_2) \\
\vdots & \vdots & \ddots & \vdots \\
\sum_{n=1}^{N} f_n(t_1) \sin(\lambda_n x_{N_x}) & \sum_{n=1}^{N} f_n(t_2) \sin(\lambda_n x_{N_x}) & \dots & \sum_{n=1}^{N} f_n(t_{N_t}) \sin(\lambda_n x_{N_x})
\end{bmatrix}
$$
Which can be rewritten as:
$$
\begin{bmatrix}
\sin(\lambda_1 x_1) & \sin(\lambda_2 x_1) & \dots & \sin(\lambda_N x_1) \\
\sin(\lambda_1 x_2) & \sin(\lambda_2 x_2) & \dots & \sin(\lambda_N x_2) \\
\vdots & \vdots & \ddots & \vdots \\
\sin(\lambda_1 x_{N_x}) & \sin(\lambda_2 x_{N_x}) & \dots & \sin(\lambda_N x_{N_x})
\end{bmatrix}
\cdot
\begin{bmatrix}
f_1(t_1) & f_1(t_2) & \dots & f_1(t_{N_t}) \\
f_2(t_1) & f_2(t_2) & \dots & f_2(t_{N_t}) \\
\vdots & \vdots & \ddots & \vdots \\
f_N(t_1) & f_N(t_2) & \dots & f_N(t_{N_t})
\end{bmatrix}
$$

We also find that:
$$\begin{bmatrix}
\sin(\lambda_1 x_1) & \sin(\lambda_2 x_1) & \dots & \sin(\lambda_N x_1) \\
\sin(\lambda_1 x_2) & \sin(\lambda_2 x_2) & \dots & \sin(\lambda_N x_2) \\
\vdots & \vdots & \ddots & \vdots \\
\sin(\lambda_1 x_{N_x}) & \sin(\lambda_2 x_{N_x}) & \dots & \sin(\lambda_N x_{N_x})
\end{bmatrix}=
\sin
\begin{bmatrix}
\lambda_1 x_1 & \lambda_2 x_1 & \dots & \lambda_N x_1 \\
\lambda_1 x_2 & \lambda_2 x_2 & \dots & \lambda_N x_2 \\
\vdots & \vdots & \ddots & \vdots \\
\lambda_1 x_{N_x} & \lambda_2 x_{N_x} & \dots & \lambda_N x_{N_x}
\end{bmatrix} = 
\sin(
\begin{bmatrix}
    \lambda_1 \\
    \lambda_2 \\
    \vdots \\
    \lambda_N
\end{bmatrix}
\begin{bmatrix}
    x_1 & x_2 & \dots & x_{N_x}
\end{bmatrix})
$$
Therefore it can be again rewritten as:
$$A = \sin(
\begin{bmatrix}
    \lambda_1 \\
    \lambda_2 \\
    \vdots \\
    \lambda_N
\end{bmatrix}
\begin{bmatrix}
    x_1 & x_2 & \dots & x_{N_x}
\end{bmatrix})
\cdot
\begin{bmatrix}
f_1(t_1) & f_1(t_2) & \dots & f_1(t_{N_t}) \\
f_2(t_1) & f_2(t_2) & \dots & f_2(t_{N_t}) \\
\vdots & \vdots & \ddots & \vdots \\
f_N(t_1) & f_N(t_2) & \dots & f_N(t_{N_t})
\end{bmatrix} = AB
$$
(A & B for purposes of clarity in python solution)

```python
def waveq_dam_solve(N, t_vec, x_vec):
    '''
     Finds the solution of the PDE
    --------------------------------------------------------------------------
    Inputs: 
        ‚Ä¢ N - int; the total number of terms in the sine series of the solution
        ‚Ä¢ t_vec - array; vector of times at which you have the solution
        ‚Ä¢ x_vec - array; vector of points in space
    Output:
        ‚Ä¢ u_x_t - array; solution u(x,t) of the PDE
    '''
    #Find the lambda values
    lambda_vals = np.arange(1, N+1)*np.pi/L
    
    #Find the sin Matrix
    A = np.sin(np.outer(lambda_vals, x_vec))
    
    #Find B
    B = np.array([fn_solve(i, t_vec) for i in range(1, N+1)])

    return (A.T@B).T

#Define N and the x and t values
x_vec = np.linspace(0, L, 201)
t_vec = np.linspace(0, 100, 1001)
N=7

u_x_t = waveq_dam_solve(N, t_vec, x_vec)
```

$$ 
  0 = -\sigma x + \sigma y, \qquad
  0 = x(\rho - z) - y, \qquad
  0 = xy - \beta z,
$$


We rearrange for x in the first equality:
$$ x = \frac{\sigma y}{\sigma} = y $$


Into the second one, we substitute y = x:
$$  0 = x(\rho - z) - x $$
ie. 
$$  0 = x(\rho - 1 - z) $$

Therefore x or ($\rho$ - 1 - z) = 0

If $x=0$:

$y = x$, so $y=0$

and $  0 = xy - \beta z$, so $0 = 0 - \beta z$, getting us $z = 0$ too.

By this we have critical point $(0, 0, 0)$.


If $(\rho - 1 - z) = 0$:

Rearranging for z:
$$  z = \rho - 1 = 12-1 = 11$$

Inserting that into the third equality, and using $y=x$:
$$
  0 = xx - \beta (\rho - 1) = x^2 - 2*11 \Rightarrow y = x = \pm\sqrt{22}
$$

Therefore the critical points are:

- $(0, 0, 0)$
- $(-\sqrt{22}, -\sqrt{22}, 11)$
- $(\sqrt{22}, \sqrt{22}, 11)$

```python
# Task 2.2 Code

#Import modules
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


#Define parameter values
sigma = 8.0
beta = 2.0
rho = 12.0  


def Lorenz(XYZ, t):
    '''
    Puts together dx/dt, dy/dt, dz/dt for scipy integration
    -------------------------------------------------------
    Inputs: 
        ‚Ä¢ XYZ - array; represents the current system [x, y, z].
        ‚Ä¢ t - float; time parameter
    Output:
        ‚Ä¢ [dxdt, dydt, dzdt] - array; computed derivatives 
    '''
    #Defines the system
    dxdt = -sigma * XYZ[0] + sigma * XYZ[1]
    dydt = XYZ[0] * (rho - XYZ[2]) - XYZ[1]
    dzdt = XYZ[0] * XYZ[1] - beta * XYZ[2]
    
    return [dxdt, dydt, dzdt]


def Plot(ax1, ax2, lab1, lab2):
    '''
    Plots a plane of ax2 against ax1
    ---------------
    Inputs:
        ‚Ä¢ ax1 - array; serves as the x-axis
        ‚Ä¢ ax2 - array; serves as the y-axis
        ‚Ä¢ lab1 - string; label for the x-axis
        ‚Ä¢ lab2 - string; label for the y-axis
    Output:
        ‚Ä¢ Plot of ax2 on ax1
    '''
    fig, ax = plt.subplots(figsize=(8, 6))

    #Plots ax2 against ax1
    ax.plot(ax1, ax2, color="black")
    #Plots final values in red
    fnl1, fnl2 = ax1[-1], ax2[-1]
    ax.scatter(fnl1, fnl2, color="red", zorder=2, label = f"Initial point destination‚âà({round(fnl1, 2)}, {round(fnl2, 2)})")


    # PLOT AESTHETICS
    #Grid
    plt.grid(True, linestyle='--')

    #Axes labels
    ax.set_xlabel(lab1)
    ax.set_ylabel(lab2)
    
    #Legend and title
    ax.legend()
    plt.title(f"{lab1}-{lab2} plane")
    
    return plt.show()

    
#Define initial values
x0, y0, z0 = 10.0, 10.0, 10.0

#Choose time step size and end time
dt = 0.00001
end = 20.0

#Generate times tested
t = np.arange(0, end+dt, dt)

#Uses scipy odeint to find solutions for ODEs
sol = odeint(Lorenz, [x0, y0, z0], t)
x, y, z = sol[:, 0], sol[:, 1], sol[:, 2]

#Plots X-Y
Plot(x, y, "x", "y")
#Plots Y-Z
Plot(y, z, "y", "z")
#Plots X-Z
Plot(x, z, "x", "z")
```

```python
# Forward Euler Scheme
def ode_Euler(xyz0, dt, end=0.3):
    '''
    Integrates the system of y' = func(y, t) using forward Euler method
    for the time steps in times and given initial condition y0
    --------------------------------------------------
    inputs:
        ‚Ä¢ xyz0 - list; initial conditions
        ‚Ä¢ dt - float; time step size
        ‚Ä¢ end - float; end time set to 0.3 automatically
    output:
        ‚Ä¢ xyz - array; solutions for every time
    '''
    #Generate times tested
    t = np.arange(0, end+dt, dt)
    
    #Initializes first row with xyz0
    xyz = [xyz0]

    #Loops for all times
    for i in range(t.size-1):
        #Solves for each time
        xyz.append(xyz[i] + dt*np.array(Lorenz(xyz[i], 0)))

    return np.array(xyz)


#Runge-Ketta method
def ode_RK4(xyz0, dt, end=0.3):
    '''
    integrates the system of y' = func(y, t) using Runge-Kutta 4th order
    for the time steps in times and given initial condition y0
    ----------------------------------------------------------
    inputs:
        ‚Ä¢ xyz0 - list; initial conditions
        ‚Ä¢ dt - float; time step size
        ‚Ä¢ end - float; end time set to 0.3 automatically
    output:
        ‚Ä¢ xyz - array; solutions for every time
    '''
    #Generates times tested
    times = np.arange(0, end + dt, dt)
    
    #Initializes position array
    xyz = np.zeros([times.size, 3])
    xyz[0, :] = xyz0
    
    #Loops for each time
    for i in range(times.size - 1):
        #Calculates each position in time using Runge-Ketta
        f1 = np.array(Lorenz(xyz[i,:], 0))
        f2 = np.array(Lorenz(xyz[i,:] + dt*f1/2, 0))
        f3 = np.array(Lorenz(xyz[i,:] + dt*f2/2, 0))
        f4 = np.array(Lorenz(xyz[i,:] + dt*f3, 0))
        xyz[i+1,:] = xyz[i,:] + dt*(f1+2*f2+2*f3+f4)/6

    return xyz


#Define initial values
xyz0 = np.array([100.0, 100.0, 100.0])

#"Exact" Runge-Ketta solution at final time
exact = ode_RK4(xyz0, 10**(-6))[-1,0]

#Time step values to test
dt_val = [0.002, 0.001, 0.0005, 0.0002, 0.0001]

#Initialize lists to store errors
EULerr = []
RK4err = []

for dt in dt_val:
    EULerr.append(np.absolute(ode_Euler(xyz0, dt)[-1, 0]-exact))
    RK4err.append(np.absolute(ode_RK4(xyz0, dt)[-1, 0]-exact)) 

fig, ax = plt.subplots(figsize=(8, 6))

# Plots the errors
ax.plot(dt_val, EULerr, 'rx', label="Forward Euler")
ax.plot(dt_val, RK4err, 'o', label="Runge-Kutta")

# PLOT AESTHETICS

#Axes labels
ax.set_xlabel("dt")
ax.set_ylabel("error")

#Legend and title
ax.legend()
plt.title("Forward Euler vs Runge-Kutta global error vs timestep")
    
#Changes to log-scale
plt.xscale('log')
plt.yscale('log')

#Grid
ax.grid(True, linestyle='--')
    
plt.show()
```

Euler has a slope of 1, while RK4 has a slope of 4 for their global error vs time step.The RK4 method is of higher order, so it converges faster to the exact solution with dt decreasing.
I would expect the slopes of the local error to be 2 and 5 respectively, which would be different from these observed ones for the global. This is because the local error is usually an order of magnitude higher. It converges faster as it takes each step's
difference into account, rather than the accumulation of "error".

```python
def plotPhase(epsilon, ogSol, dt=10**(-4), end = 60):
    '''
    Calculates the solution with x‚ÇÄ=Œµ+1, and plots the phase diagram, then the diagram of 
    the difference in solution through time with x‚ÇÄ=1
    -------------------------------------------------------
    Inputs: 
        ‚Ä¢ Œµ, epsilon - float; x‚ÇÄ=Œµ+1
        ‚Ä¢ ogSol - array; solution for [1,1,1]
        ‚Ä¢ dt - float; step size automatically set to 10**(-4)
        ‚Ä¢ end - float; end of the time tested
    Output:
        ‚Ä¢ Phase diagram, and difference vs time plot
    '''
    #Find new initial conditions
    xyz0new = np.array([1 + epsilon, 1, 1])

    #Solve for the new initial conditions
    newSol = ode_RK4(xyz0new, dt, end)
    
    #Plot the phase diagram
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(projection='3d')
    ax.plot(newSol[:,0], newSol[:,1], newSol[:,2], label = f"Initial conditions = [1+{epsilon}, 1, 1]")
    
    #Sets axis names
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    
    #Legend and title
    ax.legend()
    plt.title("Lorenz with different x‚ÇÄ=Œµ+1")
    
def plotDif(epsilon, ogSol, dt=10**(-4), end = 60):
    '''
    Calculates the solution with x‚ÇÄ=Œµ+1, and plots the diagram of 
    the difference in solution through time with x‚ÇÄ=1
    -------------------------------------------------------
    Inputs: 
        ‚Ä¢ Œµ, epsilon - float; x‚ÇÄ=Œµ+1
        ‚Ä¢ ogSol - array; solution for [1,1,1]
        ‚Ä¢ dt - float; step size automatically set to 10**(-4)
        ‚Ä¢ end - float; end of the time tested
    Output:
        ‚Ä¢ Phase diagram, and difference vs time plot
    '''
    #Find new initial conditions
    xyz0new = np.array([1 + epsilon, 1, 1])

    #Solve for the new initial conditions
    newSol = ode_RK4(xyz0new, dt, end)
    
    #Initialize difference array
    e = []
    
    #Loop for every row
    for i in range(len(newSol)):
        #Get the difference of every t
        e.append((np.sum((newSol[i]-ogSol[i])**2))**(1/2))

    #Plot the difference
    plt.figure(figsize=(10, 8))
    plt.plot(np.arange(0, end + dt, dt), e)
    
    #Axis label
    plt.xlabel('t')
    
    # Grid
    ax.grid(True, linestyle='--')
    
    #Title
    plt.title(f"Difference in solution with [1,1,1] & [1+{epsilon},1,1]")

    return plt.show()

#Original initial conditions
xyz0 = np.array([1,1,1])
#Original solution
ogSol = ode_RK4(xyz0, 10**(-4), t)
    
#Defines epsilon values
eps_vals = [1e-1, 1e-5, 1e-12]

#Loops for each epsilon
for epsilon in eps_vals:
    #Plots the phase diagram and difference plot
    plotPhase(epsilon, ogSol)
    plotDif(epsilon, ogSol)
```

When the two initial conditions are close, the two trajectories seem to stick closely together in time until a certain time, which is earlier for bigger epsilon. At that time, the difference suddenly rises and keeps dramatically changing in size, in small amounts of time. 
The difference between both trajectories seems to have an upperbound of a bit more than 50. This belief can be tested by increasing the difference in initial conditions, though I am not sure if it holds for all differences. It implies that despite difference in initial conditions, the system only allows the paths to stray slightly, having to follow the same pattern irregardless.
